```python
# Positional arguments

def print_details(*args):
    for i in args:
        print(f"positional argument is {i}")
```

```python
print_details(1,2,3,4,5,"Ram")
```

```python
# Keyword arguments

def testkeyword(**kwargs):
    for key, value in kwargs.items():
        print(f"{key} : {value}")
```

```python
testkeyword(name= "Amu", Age= 30, city= "Puna")
```

```python
# Combine Both positional and keyword arguments

def print_detail(*args, **kwargs):
    for i in args:
        print(f"positional argument is {i}")

    for key, value in kwargs.items():
        print(f"{key} : {value}")

```

```python
print_detail(1,2,6,"zaiver",name= "Amu", Age= 30, city= "Puna")
```

```python
def is_strong_passsword(password):
    """This functions checks if the password is strong or weak"""

    if len(password) < 8:
        return False

    if not any(char.isdigit() for char in password):
        return False
    
    if not any(char.islower() for char in password):
        return False

    if not any(char.isupper() for char in password):
        return False

    if not any(char in '!@#$%^&*()/*-+' for char in password):
        return False

    return True

print(is_strong_passsword("bdsiufg92364989465%%"))
```

```python
# Check if a string is a palandrome

def is_palandrome(word):
    """This function checks if the given string is palandrome"""
    
    word= word.lower().replace(" ","")
    if word == word[::-1]:
        print("It is Palandrome")
    else:
        print("It is not palandrome")

is_palandrome("Was it a car or a cat I saw")
```

```python
# Calculate the factorial of given number using recursion

def factorial(n):
    """This function prints the factorial of given number"""

    if n ==0:
        return 1
    else:
        return n * factorial(n-1)
```

```python
factorial(4)
```
