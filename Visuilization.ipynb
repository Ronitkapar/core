```python
import matplotlib.pyplot as plt
```

```python
#Creating a basic line Graph

x=[x for x in range(1,20,2)]
y=[y for y in range(21,40,2)]
print(x)
print(y)
```

```python
plt.plot(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Basic Line Graph')
plt.show()
```

```python
#Creating a Custom line Graph
y=[27,35,49,59,67,72,81,96]
x=[5,10,15,20,25,30,35,40]
plt.plot(x,y, color='red',linestyle=':')
```
