```python
#A15
def MaxMinSum(tupple):
    sort = sorted(tupple)
    print(f"The minimum value in Tupple is {sort[0]}")
    print(f"The Maximum value in Tupple is {sort[len(tupple)-1]}")

    sum = 0
    for i in tupple:
        sum += i

    print(f"The sum of the element in tupple is {sum}")

tuppllee = (14,44,55,66,77,11,55,46)
MaxMinSum(tuppllee)
        
```

```python
#A15
def MaxMinSum(tupple):
    sort = sorted(tupple)
    print(f"The minimum value in Tupple is {sort[0]}")
    print(f"The Maximum value in Tupple is {sort[len(tupple)-1]}")

    print(f"The sum of elements of tupple is {sum(tupple)}")

tuppllee = (14,44,55,66,77,11,55,46)
MaxMinSum(tuppllee)
        
```
