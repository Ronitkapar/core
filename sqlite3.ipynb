```python
import sqlite3
```

```python
#connect to a sqlite3 database
connection = sqlite3.connect('example.db')
```

```python
cursor = connection.cursor()
```

```python
cursor.execute(''' create table if not exists employees
               (id integer primary key,
               name text not null,
               age integer,
               department text)
               ''')

connection.commit()
```

```python
# Insert data into database

cursor.execute('''insert into employees
               (name,age,department)
               values('bob',45,'mechanical')''')
connection.commit()
```

```python
cursor.execute('''insert into employees
               (name,age,department)
               values('krish',22,'ai')''')
connection.commit()
```

```python
cursor.execute('''insert into employees
               (name,age,department)
               values('ronit',22,'computer')
               ''')
connection.commit()
```

```python
#Quert the data from the table
cursor.execute('select * from employees')
row=cursor.fetchall()
```

```python
# print the query data
for r in row:
    print(r)
```

```python
#update the data into the table

cursor.execute('''
               UPDATE employees
               set age=44
               where name = "krish"
               ''')
```

```python
cursor.execute('''update employees
               set department="ai"
               where name="ronit"
               ''')
```

```python
cursor.execute('select * from employees')
row=cursor.fetchall()

for r in row:
    print(r)
```

```python
#delete the data from the table
cursor.execute('''
delete from employees
               where name="bob"
               ''')
connection.commit()
```
